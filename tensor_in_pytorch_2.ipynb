```python
#to get round off value
d=torch.tensor([2.31,1.32,4.01])
torch.round(d)
```

```python
#floor pichla number pe la jata ha
torch.floor(d)
```

```python
#clamp aik range ma la ata ha numbers ko
torch.clamp(d,min=2, max=4)
```

```python
e=torch.randint(size=(2,3),low=2,high=10 ,dtype=torch.float32)
print(e)
```

```python
#all numbers are added
torch.sum(e)
```

```python
#to add rows
torch.sum(e,dim=0)
```

```python
#to add columns
torch.sum(e,dim=1)
```

```python

torch.mean(e)
```

```python
#matrix operation
g=torch.randint(size=(2,2),low=0,high=10)
h=torch.randint(size=(2,2),low=0,high=10)
print(g)
print(h)
```

```python
#matrix multiplication
torch.matmul(g,h)
```

```python
#to get dot product
vec1=torch.tensor([1,2,3])
vec2=torch.tensor([4,5,6])
torch.dot(vec1,vec2)
```

```python
#to transpose number
torch.transpose(g,0,1)
```
