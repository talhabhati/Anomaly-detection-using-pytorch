```python
import torch
```

```python
#yeh gradient true kya ha ba dma x ka derivative lena par sakta ha
x=torch.tensor(3.0, requires_grad=True)
```

```python
y=x**2
```

```python
print(x)
print(y)
```

```python
y.backward()
```

```python
#yeh derivative aya ha
x.grad
```

```python
#second example
x=torch.tensor(3.0 , requires_grad=True)
y=x**2
z=torch.sin(y)
```

```python
print(x)
print(y)
print(z)
```

```python
#derivative k liye dz/dx=(dy/dx).(dz/dy)
z.backward()
x.grad
```

```python
x.grad
```
